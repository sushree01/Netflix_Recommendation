```python
#Converting the show id column into an integer column by splitting it and extracting the numeric part
```

```python
def str_split(x):
    return re.findall('(\d+|[A-Za-z_]+)',x)
```

```python
df['show_id']=df.show_id.apply(lambda x:str_split(x)).apply(lambda x:int(x[-1]))
```

```python
df.info()
```

```python
selected_features=['type','title','director','country','rating','duration','listed_in','description']
```

```python
for i in selected_features :
    df[i]=df[i].fillna('')
```

```python
#Combining all the features
```

```python
combined_features=df['type']+' '+df['title']+' '+df['director']+' '+df['country']+' '+df['rating']+' '+df['duration']+' '+df['listed_in']+' '+df['description']
```

```python
combined_features[0]
```

```python
#Converting text data into feature vectors
```

```python
vectorizer=TfidfVectorizer()
```

```python
feature_vectors=vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
#Cosine Similarity
```

```python
#Getting the similarity score using cosine similarity
```

```python
similarity=cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
#Getting a movie name from user
```

```python
name=input('Enter your favorite movie or TV show name: ')
```

```python
#Creating a list with all the titles in the dataset
```

```python
list_of_all_titles=df['title'].to_list()
```

```python
print(list_of_all_titles)
```

```python
#Finding close match for movie name entered by the user
```

```python
find_close_match=difflib.get_close_matches(name,list_of_all_titles)
```

```python
print(find_close_match)
```

```python
#finding the closest match
```

```python
close_match=find_close_match[0]
```

```python
print(close_match)
```

```python
#finding the index of the movie with title
```

```python
index_of_the_show=df[df.title==close_match]['show_id'].values[0]
print(index_of_the_show)
```

```python
#getting a list of similar shows
```

```python
similarity_score=list(enumerate(similarity[index_of_the_show]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
#sorting the movies based on the similarity score
```

```python
sorted_similar_shows=sorted(similarity_score,key=lambda x:x[1],reverse=True)
print(sorted_similar_shows)
```

```python
#getting the names of the shows based on the index
```

```python
print('Recommended shows for you are: \n')

i=1

for show in sorted_similar_shows:
    index=show[0]
    title_from_index=df[df.index==index]['title'].values[0]
    if(i<20):
        print(i,'.',title_from_index)
        i+=1
```

We will collate all the required steps from above to design our Movie Recommendation system

```python
#Movie Recommendation System
```

```python
name=input('Enter your favorite movie or TV show name: ')

list_of_all_titles=df['title'].to_list()

find_close_match=difflib.get_close_matches(name,list_of_all_titles)

close_match=find_close_match[0]

index_of_the_show=df[df.title==close_match]['show_id'].values[0]

similarity_score=list(enumerate(similarity[index_of_the_show]))

sorted_similar_shows=sorted(similarity_score,key=lambda x:x[1],reverse=True)

print('Recommended shows for you are: \n')

i=1

for show in sorted_similar_shows:
    index=show[0]
    title_from_index=df[df.index==index]['title'].values[0]
    if(i<6):
        print(i,'.',title_from_index)
        i+=1
```
